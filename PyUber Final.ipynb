```python
# Import our dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
# Load in city data file
city_file = "raw_data/city_data.csv"
```

```python
# Load in ride data file
data_file = "raw_data/ride_data.csv"
```

```python
#Read City File
city_df = pd.read_csv(city_file)
```

```python
#drop one duplicate record with same city name, same ride_ids but different # of drivers
city_df = city_df.drop_duplicates('city', keep = 'first')
```

```python
#Read Data File
data_df = pd.read_csv(data_file)
```

```python
#merge 'city' data
merged_df = city_df.merge(data_df, on = 'city')
```

```python
#group be city name
by_city = merged_df.groupby('city')
```

```python
#Average Fare ($) Per City
avg_fare = by_city['fare'].mean()
avg_fare
```

```python
#Total Number of Rides Per City
total_rides = by_city['ride_id'].count()
total_rides
```

```python
# num of drivers
total_drivers = by_city['driver_count'].mean()
total_drivers
```

```python
# city type data
city_type = city_df.set_index('city')['type']
```

```python
#Create Dataframe for Ride Sharing Data
city_info = pd.DataFrame({
    "Number of Rides": total_rides,
    "Average Fare": avg_fare,
    "Number of Drivers": total_drivers,
    "Type of City": city_type 
})
city_info.sort_values('Number of Rides', ascending = False)
```

```python
#color scheme dictionary and preview
color_scheme = {'Gold':'#FFD700', 'Light Sky Blue':'#87CEFA', 'Light Coral':'#F08080'}
```

```python
#City Type (Urban, Suburban, Rural)
rural = city_info[city_info['Type of City'] == 'Rural']
suburban = city_info[city_info['Type of City'] == 'Suburban']
urban = city_info[city_info['Type of City'] == 'Urban']
```

```python
#Color Dictionary 
city_color = {'Urban': color_scheme['Light Coral'], 'Suburban': color_scheme['Gold'], 'Rural': color_scheme['Light Sky Blue']}
```

```python
#Scatter plot for previous data
plt.suptitle('Pyber Data Information')

# three scatter plots for each city type
plt.scatter(rural['Number of Rides'], rural['Average Fare'], s = rural['Number of Drivers']*10, color = city_color['Rural'], edgecolor = 'black', label = 'Rural', alpha = .75)

plt.scatter(suburban['Number of Rides'], suburban['Average Fare'], s = suburban['Number of Drivers']*10, color = city_color['Suburban'], edgecolor = 'black', label = 'Suburban', alpha = .75)

plt.scatter(urban['Number of Rides'], urban['Average Fare'], s = urban['Number of Drivers']*10, color = city_color['Urban'], edgecolor = 'black', label = 'Urban', alpha = .75)

#print scatter plot
plt.title('Number of Rides with Cost Per City')
plt.xlabel('Number of Rides per City')
plt.ylabel('Average Fare per City')



lgnd = plt.legend(frameon = True, edgecolor = 'black')
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]








```

```python
#% of Total Fares by City Type

#group by type to be used by all pie charts
by_type = merged_df.groupby('type')['type', 'fare', 'ride_id', 'driver_count']

#total fare by city
fare_total = by_type.sum()['fare']

#create pie chart
labels = fare_total.index

#colors and explode the same for all pie charts, reference here
colors = [city_color[n] for n in labels]
explode = [0 , 0, .2]
plt.pie(fare_total, startangle = 135, colors = colors, explode = explode, labels = labels, autopct = "%1.1f%%", shadow = True, wedgeprops = {'linewidth': .5, 'edgecolor': 'black'})

#pie chart display
plt.title('% of Total Fares by City Type')
plt.axis('equal')
plt.show()
```

```python
#% of Total Rides by City Type
 
ride_total = by_type.count()['ride_id']

labels = ride_total.index
plt.pie(ride_total, startangle = 135, explode = explode, colors = colors, labels = labels, autopct = "%1.1f%%", shadow = True, wedgeprops = {'linewidth': .5, 'edgecolor': 'black'})
plt.title('% of Total Rides by City Type')
plt.axis('equal')
plt.show()
```

```python
#% of Total Drivers by City Type
sum_driver = city_df.groupby('type').sum()['driver_count']

labels = sum_driver.index
plt.pie(sum_driver, startangle = 140, explode = explode, colors = colors, labels = labels, autopct = "%1.1f%%", shadow = True, wedgeprops = {'linewidth': .5, 'edgecolor': 'black'})
plt.title('% of Total Drivers by City Type')
plt.axis('equal')
plt.show()
```
